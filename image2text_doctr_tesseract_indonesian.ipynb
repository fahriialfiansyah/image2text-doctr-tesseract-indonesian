# docTR (Document Text Recognition)

```python
!pip install python-doctr -q
!pip install tf2onnx -q
```

```python
from doctr.io import DocumentFile
from doctr.models import ocr_predictor
```

```python
model = ocr_predictor(pretrained=True)
# Image
single_img_doc = DocumentFile.from_images("dummy-ktp.png")
# Analyze
result = model(single_img_doc)
```

```python
# Display the result (requires matplotlib & mplcursors to be installed)
result.show()
```

```python
# Extract and print the text
for page in result.pages:
    for block in page.blocks:
        for line in block.lines:
            for word in line.words:
                print(word.value, end=' ')
            print()
```

# Tesseract OCR

```python
!pip install pytesseract -q
!pip install pdf2image -q
!apt-get install poppler-utils -q
!apt-get install tesseract-ocr -q
```

```python
from PIL import Image
import pytesseract
```

```python
def extract_text_from_image(image_path):
    # open the image file
    img = Image.open(image_path)

    # use Tesseract to extract text
    text = pytesseract.image_to_string(img).strip()

    # return the extracted text
    return text

# example
image_path = 'dummy-ktp.png'
text = extract_text_from_image(image_path)
print(text)
```
